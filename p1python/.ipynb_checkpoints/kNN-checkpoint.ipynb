## Q1. Wisconsin Breast Cancer Data – Rescaling

```python
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.metrics import confusion_matrix
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import cross_val_score
```

### Load the data

```python
df = pd.read_csv('data/wbcd.csv')
```

### Explore the data

```python
df.head()
```

```python
df.describe()
```

### Split data into training and test data

```python
X = df.drop(['id', 'diagnosis'], axis = 'columns')
y = df.diagnosis

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2, random_state=5)
```

```python
clf = KNeighborsClassifier()
clf.fit(X_train, y_train)

scores = cross_val_score(clf, X_train, y_train, cv=5)
print('Validation accuracy with Unscaled Data', scores.mean())
```

```python
scaler = MinMaxScaler()
scaler.fit(X_train)
X_train_scaled = scaler.transform(X_train)

scores = cross_val_score(clf, X_train_scaled, y_train, cv=5)
print('Validation accuracy with MinMaxScaler', scores.mean())
```

### z-scaled (StandardScaler) data


```python
scalerZ = StandardScaler()
scalerZ.fit(X_train)
X_train_transformedZ = scalerZ.transform(X_train)

scores = cross_val_score(clf, X_train_transformedZ, y_train, cv=5)
print('Validation accuracy with StandardScaler', scores.mean())
```

**Best scaler would be MinMaxScaler because it has a higher accuracy**

```python
scaler = MinMaxScaler()
scaler.fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)

clf = KNeighborsClassifier()
clf.fit(X_train_scaled, y_train)

print('Test Accuracy', clf.score(X_train_scaled, y_train))
yhat = clf.predict(X_test_scaled)
cm = confusion_matrix(y_test, yhat)
print(cm)
```

## Q2. WBCD – Finding k

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, random_state=1)
scaler = MinMaxScaler()
scaler.fit(X_train)
X_train_scaled = scaler.transform(X_train)

validation_accuracy = []
training_accuracy = []
for k in range(1,15) :
    clf = KNeighborsClassifier(n_neighbors=k)
    clf.fit(X_train_scaled, y_train)
    scores = cross_val_score(clf, X_train_scaled, y_train, cv=5)
    training_accuracy.append(clf.score(X_train_scaled, y_train))
    validation_accuracy.append(scores.mean())

plt.plot(range(1,15), training_accuracy , label="training accuracy")
plt.plot(range(1,15), validation_accuracy, label="validation accuracy")
plt.xlabel("n_neighbors")
plt.ylabel("Accuracy")
plt.legend()
plt.title('Overfitting with Small value of k')
ax = plt.gca()
ax.invert_xaxis()
plt.savefig('plots/overfitting.png')
```
