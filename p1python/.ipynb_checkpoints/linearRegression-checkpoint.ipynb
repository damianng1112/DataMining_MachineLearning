## Q1: Simple Linear Regression

```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```python
# define the data
x = np.array([1, 2, 3, 4, 5, 6]).reshape(-1, 1)
y = np.array([6,1,9,5,17,12])
print('x.shape ', x.shape)
print('y.shape ', y.shape)
```

## Exploring the data

```python
plt.scatter(x, y, color = 'black')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

**Some corelation between x and y but not a lot. So don't expect a good model**

## Build the Model

```python
model = LinearRegression()
model.fit(x,y)
print('intercept:', model.intercept_)
print('slope:', model.coef_)
print('Coefficient of Determination R^2:', model.score(x,y))
```

**Explanation: About 50% of the variation in y is explained by the model**

```python
y_hat = model.predict(x)

plt.scatter(x,y, color = 'black', label = 'actual')
plt.scatter(x,y_hat, color = 'blue', label = 'predicted')
plt.plot(x,y_hat, color = 'blue', linewidth = 3)
plt.legend(loc = 'upper left')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Predicted and Actual Values of y')
plt.show()
```

**Two smallish errors in the prediction and four larger ones**

## Q2: Calculate Linear Regression Parameters

```python
x = np.array([1, 2, 3, 4, 5, 6])
y = np.array([6,1,9,5,17,12])
n = len(x)
a = ((n*sum(x*y) - (sum(x) * sum(y)))/((n*sum(x**2)- (sum(x)**2))))
print('slope:', a)
b =  ((sum(y))-(a*sum(x)))/n
print('intercept:', b)
```

## Q4. Linear Regression – Stackloss
### Objective: Predict the value of stackloss from the 3 features

```python
import pandas as pd
from pandas.plotting import scatter_matrix
import matplotlib.pyplot as plt
df = pd.read_csv("data/stackloss.csv")
df.head()
```

## Explore the data

```python
df.describe()
```

**wide spread of stackloss values, small number of very high values**

```python
df.corr()
```

**Stackloss has good correlation with AirFlow and WaterTemp but not much with AcidConc. Expect a reasonably good model**

```python
scatter_matrix(df)
plt.show()
#plt.show() help clean up the axis legend
```

**See the good correlation with AirFlow and WaterTemp. Not so much with AcidConc.**

### Prepare the data

```python
#X = df[['AirFlow', 'WaterTemp', 'AcidConc']]
X = df.drop(columns = 'StackLoss')
#y = df['StackLoss']
y = df.StackLoss
print('X.shape:',X.shape)
print('y.shape:',y.shape)
```

```python
#divide into training and testing data, random_state=1 ensures we get the same split everytime we execute
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
print('X_train.shape:',X_train.shape)
print('X_test.shape:',X_test.shape)
print('y_train.shape:',y_train.shape)
print('y_test.shape:',y_test.shape)
```

### Build the model

```python
#build the model/train the model/fit the model to the training data
model = LinearRegression()
model.fit(X_train, y_train)
print('intercept:', model.intercept_)
print('slope:', model.coef_)
print('coefficient of determination R^2:', model.score(X, y))
```

**90% of the variation in y is explained by the model, good model**
+ Equation is stackloss = (0.8\*airflow) + (1.2\*watertemp) + (-0.25*acidconc) - 36.7(the y intercept)

#### Calculate the Root Mean Square Error(RMSE) on the test data

```python
#squared=True is MSE, squared=False is RMSE
yhat = model.predict(X_test)
print(mean_squared_error(y_test, yhat, squared=False))
```

**Make Prediction for new Data**

```python
newData = [[72,20,85],
           [75,25,80]]

#create a dataframe for the new data
newDF = pd.DataFrame(newData, columns = ['AirFlow', 'WaterTemp', 'AcidConc'])
newDF.head()
```

```python
y_hat = model.predict(newDF)
print(y_hat)
```

**Airflow pushes up by 2.4(3x0.8), WaterTemp pushes up by 6(5x1.2), AcidConc pushes up by 1.25(-5x-.24)**
+ Airflow from 72 went up to 75 (3 points), the slope for airflow and stackloss from the model is approx 0.8
+ WaterTemp from 20 went up to 25 (5 points), the slope for watertemp and stackloss from the model is approx 1.2
+ AcidConc from 85 went down to 80 (5 points), the slope for acidconc and stackloss from the model is approx -0.25
