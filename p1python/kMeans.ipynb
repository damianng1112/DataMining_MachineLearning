## Q1 Iris data

```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data/iris.csv')
df.head()
```

```python
X = df.drop(columns='species')
X.head()
```

```python
model = KMeans(n_clusters=3, init='random', n_init=1, random_state=0)
cNum = model.fit_predict(X)
print(model.inertia_)
print(cNum)
```

```python
X = X.to_numpy()
plt.scatter(
 X[cNum == 0, 0], X[cNum == 0, 1],
 c='lightgreen', marker='s', edgecolor='black',
 label='cluster 1'
)
plt.scatter(
 X[cNum == 1, 0], X[cNum == 1, 1],
 c='orange', marker='o', edgecolor='black',
 label='cluster 2'
)
plt.scatter(
 X[cNum == 2, 0], X[cNum == 2, 1],
 c='lightblue', marker='v', edgecolor='black',
 label='cluster 3'
)
plt.scatter(
 model.cluster_centers_[:, 0], model.cluster_centers_[:, 1],
 s=250, marker='*',
 c='red', edgecolor='black',
 label='centroids'
)
plt.legend(scatterpoints=1)
plt.grid()
plt.xlabel("sepal_length")
plt.ylabel("sepal_width")
plt.show()
```

```python
y = df.species
ct = pd.crosstab(y, cNum)
print(ct)
```

```python
plt.scatter(
 X[cNum == 0, 2], X[cNum == 0, 3],
 c='lightgreen', marker='s', edgecolor='black',
 label='cluster 1'
)
plt.scatter(
 X[cNum == 1, 2], X[cNum == 1, 3],
 c='orange', marker='o', edgecolor='black',
 label='cluster 2'
)
plt.scatter(
 X[cNum == 2, 2], X[cNum == 2, 3],
 c='lightblue', marker='v', edgecolor='black',
 label='cluster 3'
)
plt.scatter(
 model.cluster_centers_[:, 2], model.cluster_centers_[:, 3],
 s=250, marker='*',
 c='red', edgecolor='black',
 label='centroids'
)
plt.legend(scatterpoints=1)
plt.grid()
plt.xlabel("petal_length")
plt.ylabel("petal_width")
plt.show()
```

## Q3 Cancer data

```python
df = pd.read_csv('data/wbcd.csv')
df.head()
```

```python
X = df.drop(columns=['id', 'diagnosis'])
X.head()
```

```python
scaler = StandardScaler()
X = scaler.fit_transform(X)
model = KMeans(n_clusters=2, init='random', n_init=1, random_state=0)
cNum = model.fit_predict(X)
print(model.inertia_)
print(cNum)
```

```python
plt.scatter(
 X[cNum == 0, 0], X[cNum == 0, 1],
 c='lightgreen', marker='s', edgecolor='black',
 label='cluster 1'
)
plt.scatter(
 X[cNum == 1, 0], X[cNum == 1, 1],
 c='orange', marker='o', edgecolor='black',
 label='cluster 2'
)
plt.scatter(
 model.cluster_centers_[:, 0], model.cluster_centers_[:, 1],
 s=250, marker='*',
 c='red', edgecolor='black',
 label='centroids'
)
plt.legend(scatterpoints=1)
plt.grid()
plt.xlabel("sepal_length")
plt.ylabel("sepal_width")
plt.show()
```
